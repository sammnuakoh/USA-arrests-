```python
# Let's use correlation matrix to explore this further. It refines and quantifies the correlation between the pairplots 
def plot_correlation_matrix(dataframe):
    
    
    # Selecting only numeric data
    numeric_data = dataframe.select_dtypes(include=[np.number])
    
    # Calculating the correlation matrix
    correlation_matrix = numeric_data.corr()

    # Plotting the heatmap
    sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True)
    plt.title('Correlation Matrix')
    plt.show()

plot_correlation_matrix(usarrests)
```

```python
# We choose k-means clustering over hierachical because -

# 1. it is generally faster and more computationally efficient, especially when dealing with a large dataset
# 2. it results in well-defined, non-overlapping clusters which are often more straightforward to interpret
```

```python
# We will now use Scikit Learn which is a free software machine learning library in python
from sklearn.cluster import KMeans 


# The elbow method is a technique used to determine the optimal number of clusters for a k-means clustering algorithm.
def elbow_method(dataset):
    wcss = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
        kmeans.fit(dataset)
        wcss.append(kmeans.inertia_)
    
    plt.plot(range(1, 11), wcss)
    plt.title('The Elbow Method')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS')
    plt.show()

elbow_method(usarrests)
```

```python
# Where the graph drops first indicates the optimum number of clusters
# The graph drops at 2
```

```python
# The silhouette method is another technique for determining the optimal number of clusters in a dataset, 
# particularly when using clustering algorithms like k-means.

from sklearn.metrics import silhouette_score 


def silhouette_method(dataset, kmax):
    silhouette_scores = []
    
    for k in range(2, kmax + 1):
        kmeans = KMeans(n_clusters=k).fit(dataset)
        labels = kmeans.labels_
        silhouette_scores.append(silhouette_score(dataset, labels, metric='euclidean'))

    plt.figure(figsize=(8, 6))
    plt.plot(range(2, kmax + 1), silhouette_scores, marker='o', linestyle='-', color='b')
    plt.title('Silhouette Score for Optimal K')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Silhouette Score')
    plt.grid()
    plt.show()

silhouette_method(usarrests, 10)
```

```python
# The first drop or point is the optimum number of clusters
# The first drop or point is 2

# We have the same value (k value) for both elbow and silhouette methods
```

```python
# Now using the k-means clustering method to specify number of clusters
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
```

```python
# Fitting k-means clustering method into our dataset 

kmeans.fit(usarrests) 
```

```python
usarrests['Cluster'] = kmeans.labels_
```

```python
# Import Principal Component Analysis as PCA from Scikit Learn library
# PCA helps to identify and emphasize the most important patterns and relationships in your data
# PCA can help remove noise or irrelevant information from your data
from sklearn.decomposition import PCA

# Add cluster labels to the Data Frame and name them
usarrests['Cluster'] = kmeans.labels_
usarrests['Cluster'] = usarrests['Cluster'].map({0: 'Cluster 0', 1: 'Cluster 1'})

# PCA for dimensionality reduction
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(usarrests.drop('Cluster', axis=1))

# Scatter plot with annotations
plt.figure(figsize=(10, 6))
for cluster_name in ['Cluster 0', 'Cluster 1']:
    cluster_data = reduced_data[usarrests['Cluster'] == cluster_name]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=cluster_name, marker='o')
    

plt.title('Visualization of Clusters with Labels')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.show()
```

```python
# Risk Scoring in machine learning (ML) 

import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.stats import zscore

def calculate_improved_risk_score(data, crime_columns):
    
    # Standardize the crime rates using Z-score normalization
    standardized_data = data[crime_columns].apply(zscore)

    # Handling outliers: Capping and Flooring at specific Z-scores (e.g., -3 and 3)
    capped_data = standardized_data.clip(lower=-3, upper=3)

    # Assign equal weights to all crime types
    weights = {crime: 1 / len(crime_columns) for crime in crime_columns}

    # Calculate weighted sum of crime rates for each state
    data['RiskScore'] = capped_data.apply(lambda row: sum(row[crime] * weights[crime] for crime in crime_columns), axis=1)

    # Optionally, normalize the Risk Scores to a scale of 0-100
    max_score = data['RiskScore'].max()
    data['NormalizedRiskScore'] = (data['RiskScore'] / max_score) * 100

    return data

# Example usage
crime_columns = ['Murder', 'Assault', 'Rape']
usarrests_improved = calculate_improved_risk_score(usarrests, crime_columns)
usarrests_improved[['Murder', 'Assault', 'Rape', 'RiskScore', 'NormalizedRiskScore']].head(20)

```
